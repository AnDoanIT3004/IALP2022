```python
df1= pd.read_excel('/content/drive/MyDrive/DoAnTotNghiep/Data/train_nor_811.xlsx')
df1
```

```python
train=pd.DataFrame()
train['Emotion']=df1['Emotion']
train['Sentence']=df1['Sentence']
```

```python
import re
import json 

pattern = '\s'
```

```python
re.split(pattern, str(train['Sentence'][1]))
```

```python
data_count=pd.DataFrame()
data_count['ID']= np.arange(len(train))
data_count['Disgust']= np.nan
data_count['Fear']= np.nan
data_count['Enjoyment']= np.nan
data_count['Sadness']= np.nan
data_count['Suprise']= np.nan
data_count['Anger']= np.nan
data_count['Other']= np.nan
```

```python
data_count
```

```python
for i in range(0,len(train)):
  count_disgust=0
  count_fear=0
  count_enjoyment=0
  count_sadness=0
  count_suprise=0
  count_other=0
  count_anger=0
  for j in range(0,len(data)):
    if (re.split(pattern, str(train['Sentence'][i]))).count(str(data['vietnamese'][j]))!=0:
      if data['Anger'][j]==1:
        count_disgust+=1
      if data['Fear'][j]==1:
        count_fear+=1
      if data['Enjoyment'][j]==1:
        count_enjoyment+=1
      if data['Sadness'][j]==1:
        count_sadness+=1
      if data['Suprise'][j]==1:
        count_suprise+=1
      if data['Other'][j]==1:
        count_other+=1
      if data['Disgust'][j]==1:
        count_disgust+=1
    
  data_count['Disgust'][i]= count_disgust
  data_count['Fear'][i]= count_fear
  data_count['Enjoyment'][i]= count_enjoyment
  data_count['Sadness'][i]= count_sadness
  data_count['Suprise'][i]= count_suprise
  data_count['Anger'][i]= count_anger
  data_count['Other'][i]= count_other
```
